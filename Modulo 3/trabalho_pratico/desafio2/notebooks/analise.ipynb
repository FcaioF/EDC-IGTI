```python
#imports
from pyspark.sql import SparkSession
from pyspark.sql.functions import * 
from pyspark.sql.types import *
```

```python
#instanciando spark 
spark = SparkSession.builder \
    .master('local[2]') \
    .appName('desafio') \
    .config('spark.ui.port', '4030') \
    .getOrCreate()
```

```python
#diretorios de trabalho
stage_enade = 'D://caiof//Documents//GIT_Repos//EDC-IGTI//Modulo 3//base_dados//base_desafio2//microdados_enade_2017_LGPD//stage_data//'
```

Qual era o numero de alunos de filosofia inscritos no ENADE 2017?

```python
##leitura da base de dados
enade_df = (
    spark
    .read
    .format('parquet')
    .option('header',True)
    .option('inferSchema',True)
    .option('encoding','ISO-8859-1')
    .load(f'{stage_enade}microdados2017_arq1')
)

enade_df.show()
```

```python
(
    enade_df
    .select('CO_GRUPO')
    .filter('CO_GRUPO == 3201 or CO_GRUPO == 3202')
    .count()
    # .show(15,truncate=False)
)
```

Qual o numero de alunos dos cursos de filosofia do sexo masculino

```python
enade_df_2 = (
    spark
    .read
    .format('parquet')
    .option('header',True)
    .option('inferSchema',True)
    .option('encoding','ISO-8859-1')
    .load(f'{stage_enade}microdados2017_arq5')
)
```

join entre tabelas

```python
#Junção dos dataframes em um unico (JOIN)
joined_df = (
    enade_df
    .join(enade_df_2,enade_df.NU_ANO == enade_df_2.NU_ANO,'full')
)

joined_df.show(truncate=False)
```
